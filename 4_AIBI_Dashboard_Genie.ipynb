## Genie Room

1. Add Data Sources
- review_aspect_details
- issues_daily
- raw_reviews

2. Add Descriptions
> ```
> This data room contains reivews about Lakehouse Inn
> 
> Available Data:
> - raw_reviews:  Hotel guest reviews from various channels and locations. Each row is one review.
> 
> - review_extractions: Raw JSON output from ai_query extractions containing structured review insights (aspects, entities, metadata, sentiment).
> 
> - issues_daily: Detected open issues representing statistically significant negative trends per aspect/location/day. Includes metadata, severity, and relevant reviews.
> ```

3. (Optional) Add sample questions

`When was room_quality first mentioned in a negative review?`

`Give me insights about the aspect with most negative reviews`

4. (Optional) Add Insturctions
5. Make sure to get the **Genie Room ID** from url 

`<Host>/genie/rooms/<Genie Room ID>`
