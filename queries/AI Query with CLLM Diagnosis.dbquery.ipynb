```sql
CREATE OR REPLACE TABLE IDENTIFIER(:catalog || '.' ||:schema || '.' || :output_table) AS
WITH query_results AS (
  SELECT
    *,
    ai_query(
       :cllm_agent_endpoint,
      relevant_reviews_text,
      failOnError => false
    ) AS response
  FROM (
    SELECT *
    FROM IDENTIFIER(:catalog || '.' ||:schema || '.' || :issues_table)
  )
)
SELECT
  -- Exclude the response column
  issue_id,
  opened_at,
  location,
  aspect,
  severity,
  status,
  open_reason,
  delta_pp_open,
  nms_open,
  volume_open,
  relevant_reviews,
  response.result AS response_result,
  response.errorMessage AS response_error,
  current_date() AS processed_date
FROM query_results;
```
